#### Tabla de Amortización

```python
import numpy_financial as npf
import tabulate as tab
```

```python
capital = 190000
tasa = 0.325/12
tasa_adicional = .0064                             # Tasa contratada 32.49%
tasa_total = tasa + tasa_adicional
plazo = 60

cuota = round(npf.pmt(tasa_total, plazo, -capital, 0), 0)
datos = []
saldo = capital
```

```python
for i in range(1, plazo + 1):
    pago_capital = npf.ppmt(tasa_total,i, plazo, -capital, 0)
    pago_int = cuota - pago_capital
    saldo -= pago_capital
    linea = [i,format( cuota, "0,.0f"), format(pago_capital, "0,.0f"),
            format(pago_int, "0,.0f"),format(saldo, "0,.0f")]
    datos.append(linea)

print(tab.tabulate(datos, headers = ["Periodo", "Cuota", "Capital", "Intereses","saldo"],
      tablefmt = "psql"))
```

```python
capital = 268000                          # Para calcular el IVA de los intereses,apliqué
tasa = .365                                 # el 36.5% en lugar del 29% .
plazo = 5
periodos = plazo * 12

pago = ( npf.pmt( tasa / 12, periodos, -capital),2)
tabla = []
# Iterar sobre cada periodo
for i in range(1, periodos + 1):
    interes = npf.ipmt(tasa / 12, i, periodos, -capital)
    amortizacion = npf.ppmt(tasa / 12, i , periodos, -capital)
# Calcular el saldo pendiente al final del periodo
    saldo = (capital - amortizacion,2)
# Añadir la fila correspondiente a la tabla
    tabla.append([i, pago, interes, amortizacion, saldo])
#Actualizar el capital para el siguiente periodo
    saldo -= capital

encabezados = ["Periodo","Pago","Interes","Amortización", "Saldo"]

# Imprimir la tabla usando la función tabulate
print(tabulate(tabla, encabezados,tablefmt = "fancy_grid"))
```
